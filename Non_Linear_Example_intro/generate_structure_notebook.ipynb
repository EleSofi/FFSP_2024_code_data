```python
import os

def find_project_root(starting_dir='.', root_marker='FFSP_code_revised_2024'):
    current_dir = os.path.abspath(starting_dir)
    while True:
        if root_marker in os.listdir(current_dir):
            return os.path.join(current_dir, root_marker)
        parent_dir = os.path.abspath(os.path.join(current_dir, os.pardir))
        if parent_dir == current_dir:
            raise FileNotFoundError(f"Project root directory '{root_marker}' not found.")
        current_dir = parent_dir

# Determine the path to the current notebook's directory
notebook_dir = os.getcwd()
print(f"Notebook directory: {notebook_dir}")

# Find the project root
try:
    project_root = find_project_root(notebook_dir)
    print(f"Project root found: {project_root}")
except FileNotFoundError as e:
    print(e)
    project_root = None

# Construct the path to the markdown generation script relative to the project root directory
if project_root:
    script_path = os.path.join(project_root, 'Non_Linear_Example_intro', 'generate_directory_markdown.py')
else:
    script_path = None

# Check if the path exists
if script_path:
    print(f"Script path: {script_path}")
    print(f"Exists: {os.path.exists(script_path)}")
else:
    print("Script path not determined.")

# Run the script
if script_path and os.path.exists(script_path):
    %run $script_path
else:
    print("Script not found. Please check the path.")
```

```python
# Display the generated markdown file
output_markdown_path = os.path.join(project_root, 'Non_Linear_Example_intro', 'project_structure.md')

# Read and print the contents of the markdown file
if os.path.exists(output_markdown_path):
    with open(output_markdown_path, 'r') as f:
        markdown_content = f.read()
    print(markdown_content)
else:
    print("Markdown file not found. Please check the path.")
```
