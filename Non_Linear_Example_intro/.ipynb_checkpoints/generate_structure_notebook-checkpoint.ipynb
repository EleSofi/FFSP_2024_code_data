```python
import os

# Determine the path to the current notebook's directory
notebook_dir = os.getcwd()

# Construct the path to the script relative to the project root directory
script_relative_path = os.path.join('Non_Linear_Example_intro', 'generate_directory.py')

# Walk up the directory tree to find the project root
current_dir = notebook_dir
while current_dir and not os.path.exists(os.path.join(current_dir, script_relative_path)):
    parent_dir = os.path.abspath(os.path.join(current_dir, os.pardir))
    if parent_dir == current_dir:
        break  # Reached the root of the filesystem
    current_dir = parent_dir

script_path = os.path.join(current_dir, script_relative_path)

# Check if the path exists
print(f"Script path: {script_path}")
print(f"Exists: {os.path.exists(script_path)}")

# Run the script
if os.path.exists(script_path):
    %run $script_path
else:
    print("Script not found. Please check the path.")
```

```python
# Construct the path to the output file relative to the project root directory
output_file_relative_path = os.path.join('Non_Linear_Example_intro', 'project_structure.md')
output_file = os.path.join(current_dir, output_file_relative_path)

# Read and print the contents of the output file
if os.path.exists(output_file):
    with open(output_file, 'r') as f:
        print(f.read())
else:
    print("Output file not found. Please check the path.")
```
